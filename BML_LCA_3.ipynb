```python
import pandas as pd
import numpy as np
import seaborn as sb

dt=pd.read_csv('/content/drive/MyDrive/data.csv')
dt.info()
```

```python
dt['gender']=dt['gender'].map({"Male":1,"Female":0})
dt['Partner']=dt['Partner'].map({"Yes":1,"No":0})
dt['PhoneService']=dt['PhoneService'].map({"Yes":1,"No":0})
dt['Churn']=dt['Churn'].map({"Yes":1,"No":0})
dt['DeviceProtection']=dt['DeviceProtection'].map({"Yes":1,"No":0,"No internet service":2})
dt['PaperlessBilling']=dt['PaperlessBilling'].map({"Yes":1,"No":0})
dt['OnlineSecurity']=dt['OnlineSecurity'].map({"Yes":1,"No":0,"No internet service":2})
dt['TechSupport']=dt['TechSupport'].map({"Yes":1,"No":0,"No internet service":2})
dt['StreamingTV']=dt['StreamingTV'].map({"Yes":1,"No":0,"No internet service":2})
dt['StreamingMovies']=dt['StreamingMovies'].map({"Yes":1,"No":0,"No internet service":2})
```

```python
print(dt.head())
```

```python
d=dt[['gender','SeniorCitizen','Partner','tenure','PhoneService','OnlineSecurity','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','MonthlyCharges','Churn']]
```

```python
sb.heatmap(d.corr(),annot=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
x=dt[['StreamingMovies','StreamingTV','MonthlyCharges','tenure','PaperlessBilling','Partner']]
y=dt['Churn']
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,test_size=0.2,random_state=45)
model=SVC()
model.fit(x_train,y_train)
# prediction=model.predict(x_test,y_test)
print("Model Accuracy:",model.score(x_test,y_test))
```
